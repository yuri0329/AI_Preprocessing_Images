# Image Crop

```python
def Process(image_path, label_path=[], size=(1440,1080)):
  img = cv2.imread(image_path)
  # Read image size
  y, x, ch = img.shape

  # Image resize
  img = Image_Resize(img, size)
  # Save image
  cv2.imwrite(image_path, img)

  # If labeling (if there is a txt file)
  if label_path:
    # Reading from YOLO coordinates to CV2 coordinates
    label_list = Yolo_to_CV2(y, x, size, label_path)
    # Coordinates are corrected as much as cropped and converted back to YOLO coordinates
    label = CV2_to_Yolo(size, label_list)
    # Save YOLO coordinates
    with open(label_path, 'w') as f:
        f.writelines(label)

  #return img, label
```

```python
# Image resizing
def Image_Resize(img, size):
    return cv2.resize(img, (size[0], size[1]))
```

```python
# YOLO -> CV2
def Yolo_to_CV2(dh, dw, size, label):
  import pybboxes as pbx

  # Read txt file
  with open(label, 'r') as f:
      data = f.readlines()

  label_list = []

  # YOLO -> cv2
  for dt in data:
      # Split string to float
      _, x, y, w, h = map(float, dt.split(' '))
      
      xy_cordinate = list(pbx.convert_bbox((x, y, w, h), from_type="yolo", to_type="voc", image_size=(dw, dh)))

      for idx in range(len(xy_cordinate)):
        if idx % 2:
          xy_cordinate[idx] = round(size[1]/dh * xy_cordinate[idx])
        else:
          xy_cordinate[idx] = round(size[0]/dw * xy_cordinate[idx])

      label_list.append(tuple(xy_cordinate))

  return label_list
```

```python
# Check if value is - or greater than 1
def check(cor_list):
  for item in cor_list:
    if item <= 0 or item > 1:
      return True
  return False
```

```python
# CV2 -> YOLO
def CV2_to_Yolo(size, label_list):
  import pybboxes as pbx

  text_list = []
  for box in label_list:
    print(box)
    x, y, w, h = pbx.convert_bbox(box, from_type="voc", to_type="yolo", image_size=(size[0], size[1]))

    # Don't save if value is less than 0 or greater than 1
    if check([x, y, w, h]):
      print('error2 :', [x, y, w, h])
      continue
    
    # If the box width is less than 0.05, don't save it because it's too small
    if w < 0.03:
      print('too small')
      continue
    text = '0 {} {} {} {}'.format(x, y, w, h)
    text_list.append(text)
    
  return ('\n').join(text_list)
```

# Actual code

```python
import os
import cv2 

# Images and txt file path + /
path = '' 

# Get the names of all files in a folder
file_list = os.listdir(path)

# enumerate : Returns the index (order) and two elements of the list
#  ['', '', ''] -> (0, ''), (1, ''), (2, '')
for idx, item in enumerate(file_list):
  # Divide files by name and extension

  name, ext = os.path.splitext(item)
  # if it's a image
  if(ext == '.jpg'):
    # Initialize the label path (to crop and resize unlabeled photos)
    label_path = ''
    # True if there is a txt file
    if os.path.isfile(path+name+'.txt'):
        label_path = path+name+'.txt'
    # Process function start
    Process(path+item, label_path, (1440, 1080))
```
